```python
# Basic math operations
print(10 + 5) # 15 (Addition)
print(10 - 5) # 5 (Subtraction)
print(10 * 5) # 50 (Multiplication)
print(10 / 5) # 2.0 (Division - always float)
print(10 // 3) # 3 (Integer division)
print(10 % 3) # 1 (Modulus - remainder)
print(2 ** 3) # 8 (Exponentiation)
```

```python
age = 25
has_license = True
#AND - both must be True 
can_drive = age >= 16 and has_license
print(can_drive) # True
# OR - at least one must be True
gets_discount = age < 18 or age > 65
print(gets_discount) # False
```

```python
# Combining conditions
age = 25
has_license = True
# AND - both must be True
can_drive = age >= 16 and has_license
print(can_drive)  # True
# OR - at least one must be True
gets_discount = age < 18 or age > 65
print(gets_discount)  # False
```

```python
# EX 1: VARIABLES & PRINT
# about_me.py - Answer Key
# Store a string for the name
name = "Alex"
# Store an integer for the age
age = 28
# Store a float for the height (in feet or meters)
height = 5.9
# Store a boolean for programming preference
likes_programming = True
# Display all information using professional f-strings
print(f"My name is {name}.")
print(f"I am {age} years old and {height} tall.")
print(f"Do I like programming? {likes_programming}")
```

```python
#EX 2: CALCULATOR
# calculator.py - Answer Key
# 1. Variables
num1 = 15
num2 = 4
# 2. Operations & Clarity (using f-strings)
print(f"Addition: {num1} + {num2} = {num1 + num2}")
print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
print(f"Multiplication: {num1} * {num2} = {num1 * num2}")
print(f"Standard Division: {num1} / {num2} = {num1 / num2}")
print(f"Floor Division: {num1} // {num2} = {num1 // num2}")
print(f"Modulus (Remainder): {num1} % {num2} = {num1 % num2}")
```

```python
# EX 3: AGE CHECKER
# age_checker.py - Answer Key
# 1. Logic: Initialize variables
your_age = 20
voting_age = 18
# 2. Check & Output: Perform logical comparisons
# Greater than or equal to check
can_vote = your_age >= voting_age
print(f"Am I old enough to vote? {can_vote}")
# Less than check
is_minor = your_age < 18
print(f"Am I a minor? {is_minor}")
# Equality check
is_exactly_voting_age = your_age == voting_age
print(f"Am I exactly the voting age? {is_exactly_voting_age}")
```
