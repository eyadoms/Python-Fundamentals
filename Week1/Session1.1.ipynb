```python
# Basic math operations
print(10 + 5) # 15 (Addition)
print(10 - 5) # 5 (Subtraction)
print(10 * 5) # 50 (Multiplication)
print(10 / 5) # 2.0 (Division - always float)
print(10 // 3) # 3 (Integer division)
print(10 % 3) # 1 (Modulus - remainder)
print(2 ** 3) # 8 (Exponentiation)
```

```python
age = 25
has_license = True
#AND - both must be True 
can_drive = age >= 16 and has_license
print(can_drive) # True
# OR - at least one must be True
gets_discount = age < 18 or age > 65
print(gets_discount) # False
```

```python
# Combining conditions
age = 25
has_license = True
# AND - both must be True
can_drive = age >= 16 and has_license
print(can_drive)  # True
# OR - at least one must be True
gets_discount = age < 18 or age > 65
print(gets_discount)  # False
```
